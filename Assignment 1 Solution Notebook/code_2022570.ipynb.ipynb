```python
import pandas as pd
# importing library

df = pd.read_csv("/content/train.csv")
# reading from the file

print(df.shape) # printing the shape
print(df.head(10)) # printing the first 10 entries
print(df.tail(10)) # printing the last 10 entries
```

```python
print(df.isnull().sum())
# printing the number of null values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# importing libraries

sns.countplot(x = "Survived", data = df)
plt.show()
# ploting the graph for count againt survived
sns.countplot(x = "Pclass", data = df)
plt.show()
# ploting the graph for count againt Pclass
sns.countplot(x = "Name", data = df)
plt.show()
# ploting the graph for count againt Name
sns.countplot(x = "Sex", data = df)
plt.show()
# ploting the graph for count againt Sex
sns.countplot(x = "Age", data = df)
plt.show()
# ploting the graph for count againt Age
sns.countplot(x = "SibSp", data = df)
plt.show()
# ploting the graph for count againt Sibsb
sns.countplot(x = "Parch", data = df)
plt.show()
# ploting the graph for count againt Parch
sns.countplot(x = "Embarked", data = df)
plt.show()
# ploting the graph for count againt Embarked
```

```python
df.groupby(["Sex", "Survived"])["Survived"].count()
# 
```

```python
sns.countplot(x = "Pclass", hue = "Survived", data = df)
```

```python
survived = df[df["Survived"] == 1]
print("Oldest Survivor: ", survived["Age"].max())
print("Youngest Survivor: ", survived["Age"].min())
print("Average age of survivors: ", survived["Age"].mean())
```
